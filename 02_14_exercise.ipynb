```python
# Create a function to open and read in daily mean temperature for NYC Central Park (2010) from a
# .txt file. Using the readlines command to make the output a list.
def read_text(file):
    with open(file, 'r') as data:
        sat = data.readlines()
    return sat
```

```python
# Print the observed NYC temperature to the screen
def print_data(obs):
    day_of_year = [str(k) for k in range(1, 366)]
    day_of_year.insert(0, 'Day of Year\n(1-365)')
    temp_column = ['Temperature Observation\n(Degrees Fahrenheit)'] + obs
# Zip the two lists together for easier printing
    combined_data = list(zip(day_of_year, temp_column))
    
# this null function is passed the temperature data. The function prints to the screen two columns with headers:
    # - one column indicating the day of the year (1-365),
    # - and a second column indicating the temperature observation for each day.
# Be sure to indicate the units of the temperature data (Â°F)

    for item in combined_data:
        print('\t'.join(item))
    return
```

```python
# Main Routine (Exercise 1 - 02/14)

# First, specify the file name that has already been included in the same folder:
file_text = 'nyc_temp_2010.txt'
    # this text file contains the daily mean temperatures for NYC Central Park (2010)

# Now call the functions (read and print):
sat_nyc_2010 = read_text(file_text)
    # this function is passed the file name, reads in the temperature data, and returns the data

print_data(sat_nyc_2010)
```
